```python
# 給csv檔目錄路徑
path = "songinfo/"

# 用glob將路徑內的.csv檔全部抓出來
all_files = glob(os.path.join(path, "*.csv"))

# 合併所有csv檔
# 一個一個將all_files裡的csv檔讀進來，不留index，並用逗號隔開[1csv,2csv,3csv,....]，才能用concat語法
songinfo_from_each_file = (pd.read_csv(f, sep=',') for f in all_files) 
songinfo = pd.concat(songinfo_from_each_file, ignore_index=True) # 將所有讀進來的csv檔合併，並忽略index
```

```python
songinfo.to_csv('./allsonginfo/allsonginfo.csv')
```
